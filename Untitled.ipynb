## Data cleaning


- Remove HTML tags
- Remove non-alphanumeric characters
- Convert to lowercase
- Tokenize and Remove stopwords
- Lemmatisation 

### Remove HTML tags

```python
def remove_html(text):
    pattern = re.compile(r'<.*?>')
    return re.sub(pattern,'',text)

data.review = data.review.apply(remove_html)
data.review[0]
```

### Remove non-alphanumeric characters

```python
def special_character(text):
    without_special_char = ''
    for i in text:
        if i.isalnum(): #The isalnum() method returns True if all the characters are alphanumeric, meaning alphabet letter (a-z) and numbers (0-9)
            without_special_char = without_special_char + i
        else:
            without_special_char = without_special_char + ' '
    return without_special_char

data.review = data.review.apply(special_character)
data.review[0]
```

### Convert  to lowercase


```python
data.review = data.review.apply(lambda x: x.lower())
data.review[0]
```

### Tokenize and  remove stopwords

```python
def tokenize_remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    return [word for word in tokens if word not in stop_words]

data.review = data.review.apply(tokenize_remove_stopwords)
print(data.review[0])
```

### Lemmatisation  


```python
def lemmatize_text(text):
    lemmatizer = WordNetLemmatizer()
    return " ".join([lemmatizer.lemmatize(word) for word in text])

data.review = data.review.apply(lemmatize_text)
print(data.review[0])
```

### Biogram

```python
# Chain Make an iterator that returns elements from the first iterable until it is exhausted, then proceeds to the next iterable, until all of the iterables are exhausted.
# Counter is a container that stores elements as dictionary keys, and their counts are stored as dictionary values.

lemmatized_tokens = [word.split() for word in data.review]
list_tokens = list(chain(*lemmatized_tokens))
biograms = zip(list_tokens,list_tokens[1:]) #takes iterable or containers and returns a single iterator object, having mapped values from all the containers.
dict_biograms = Counter(biograms)
df = pd.DataFrame(dict_biograms.most_common(30), 
                 columns = ['Biogram','Count'])
fig,ax = plt.subplots(figsize=(12,8))
df.sort_values(by='Count').plot.barh(x='Biogram',y='Count', ax=ax)
ax.set_title('Most Frequently used Biograms in reviews')
plt.show()
```

## Evaluation Metrics

```python
def display_confusion_matrix(true_labels, predicted_labels, classes=[0,1]):
    
    total_classes = len(classes)
    level_labels = [total_classes*[0], list(range(total_classes))]

    cm = confusion_matrix(true_labels, predicted_labels, labels=classes)
    cm_frame = pd.DataFrame(data=cm, 
                            columns=pd.MultiIndex(levels=[['Predicted:'], classes], 
                                                  codes=level_labels), 
                            index=pd.MultiIndex(levels=[['Actual:'], classes], 
                                                codes=level_labels)) 
    print(cm_frame) 
    

def model_performance_metrics(true,prediction,classes=[0,1]):
    print("Note: positive = 1, negative = 0")
    print("\n")
    print("Accuracy = ", accuracy_score(true,prediction))
    print("Precision = ", precision_score(true,prediction))
    print("Recall = ", recall_score(true,prediction))
    print("F1_score = ", f1_score(true,prediction))
    print("\n")
    print("Classification Report \n", classification_report(true,prediction))
    print("\n")
    print("Confusion Matrix \n")
    display_confusion_matrix(true,prediction, classes=classes)
                                                          
    return

def plot_precision_recall(precisions, recalls, label=None):
    plt.figure(figsize=(8, 6))                                    
    plt.plot(precisions, recalls, linewidth=2, label=label)
    plt.xlabel("Recall", fontsize=16)
    plt.ylabel("Precision", fontsize=16)
    plt.title('Precision-Recall curve')
    plt.grid(True)   

def plot_roc_curve(fpr, tpr, label=None):
    plt.figure(figsize=(8, 6))                                    
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
    plt.axis([0, 1, 0, 1])                                    
    plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16)
    plt.ylabel('True Positive Rate (Recall)', fontsize=16)    
    plt.grid(True) 
```

#### Hyperparameter tuning  - Cross-Validation

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

#Defining model and hyperparameters
model = LogisticRegression()
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = ['l2']
c_values = [100, 10, 1.0, 0.1, 0.01]

# define grid search
grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result = grid_search.fit(partial_X_train, partial_y_train)

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```

```python
#save final Logistic regression model
final_model_lr = grid_result.best_estimator_
```

#### Validation accuracy, precision, recall and f1-score - Final model (Cross-Validation)

```python
from sklearn.model_selection import cross_val_score

results = cross_val_score(final_model_lr,X_val,y_val, cv=cv, n_jobs=-1,scoring="accuracy")
accuracy = round(np.mean(results),4)

print("Accuracy:  %r" % accuracy)
```

```python
#Save accuracy from final model obtained from grid search
results_df = pd.DataFrame(data=[['Logistic Regression', accuracy]], 
                          columns=[['Model', 'Accuracy']])
results_df
```

```python
pred = cross_val_predict(final_model_lr,X_val,y_val, cv=cv, n_jobs=-1)

model_performance_metrics(y_val,pred,classes=[0,1])
```

```python
from sklearn.metrics import precision_recall_curve

precisions, recalls, thresholds = precision_recall_curve(y_val,pred)

plot_precision_recall(precisions, recalls)
plt.show() 
```

#### ROC curve 


The ROC curve shows the trade-off between sensitivity (or TPR) and specificity (1  FPR). Classifiers that give curves closer to the top-left corner indicate a better performance.

```python
y_scores = cross_val_predict(final_model_lr, X_val,y_val, cv=cv,
                             method="decision_function", n_jobs=-1)
```
